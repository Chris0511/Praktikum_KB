## Mengimport Library dan Membaca Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
dataFrame = pd.read_csv('athlete_events.csv')
dataFrame.head()
```

## Bagi dataset menjadi training set dan testing set dengan proporsi 70:30

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Baca dataset
df = pd.read_csv('athlete_events.csv')

# Bagi dataset menjadi training set dan testing set
train_set, test_set = train_test_split(df, test_size=0.3, random_state=42)

# Menampilkan beberapa baris pertama dari training set
print("Training Set:")
print(train_set.head())

# Menampilkan beberapa baris pertama dari testing set
print("\nTesting Set:")
print(test_set.head())
```

## Normalisasi data pada attribut 'Age' menggunakan Min Max scaler

```python
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Pilih atribut yang ingin Anda normalisasi (misalnya, age)
atribut_normalisasi = ['Age']

# Lakukan normalisasi pada training set
train_set[atribut_normalisasi] = scaler.fit_transform(train_set[atribut_normalisasi])

# Lakukan normalisasi pada testing set
test_set[atribut_normalisasi] = scaler.transform(test_set[atribut_normalisasi])

# Menampilkan beberapa baris pertama dari training set setelah normalisasi
print("Training Set setelah Normalisasi:")
print(train_set.head())

# Menampilkan beberapa baris pertama dari testing set setelah normalisasi
print("\nTesting Set setelah Normalisasi:")
print(test_set.head())
```

## Standarisasi pada atribut 'Age'

```python
from sklearn.preprocessing import StandardScaler

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Pilih atribut yang ingin Anda standarisasi (misalnya, age)
atribut_standarisasi = ['Age']

# Lakukan standarisasi pada training set
train_set[atribut_standarisasi] = scaler.fit_transform(train_set[atribut_standarisasi])

# Lakukan standarisasi pada testing set
test_set[atribut_standarisasi] = scaler.transform(test_set[atribut_standarisasi])

# Menampilkan beberapa baris pertama dari training set setelah standarisasi
print("Training Set setelah Standarisasi:")
print(train_set.head())

# Menampilkan beberapa baris pertama dari testing set setelah standarisasi
print("\nTesting Set setelah Standarisasi:")
print(test_set.head())
```

## Cek ada nilai Null pada Dataset

```python
# Periksa apakah ada nilai null pada training set
train_null_count = train_set.isnull().sum()

# Periksa apakah ada nilai null pada testing set
test_null_count = test_set.isnull().sum()

# Tampilkan jumlah nilai null pada setiap kolom
print("Jumlah nilai null pada training set:")
print(train_null_count)

print("\nJumlah nilai null pada testing set:")
print(test_null_count)
```

## Melakukan Data Cleaning pada nilai Null

```python
# Data Cleaning pada Age
median_age = train_set['Age'].median()
train_set['Age'].fillna(median_age, inplace=True)

# Data Cleaning pada Height
mean_height = train_set['Height'].mean()
train_set['Height'].fillna(mean_height, inplace=True)

# Data Cleaning pada Weight
mean_weight = train_set['Weight'].mean()
train_set['Weight'].fillna(mean_weight, inplace=True)

# Data Cleaning pada Medal
mode_medal = train_set['Medal'].mode()[0]
train_set['Medal'].fillna(mode_medal, inplace=True)

# Periksa apakah masih ada nilai null
train_null_count = train_set.isnull().sum()

# Tampilkan hasil
print("Jumlah nilai null pada training set setelah data cleaning:")
print(train_null_count)
```

## Cek data Duplikat pada Dataset

```python
# Periksa nilai duplikat pada training set
duplicate_rows_train = train_set[train_set.duplicated()]

# Periksa nilai duplikat pada testing set
duplicate_rows_test = test_set[test_set.duplicated()]

# Tampilkan nilai duplikat (jika ada) dari training set
print("Nilai duplikat pada training set:")
print(duplicate_rows_train)

# Tampilkan nilai duplikat (jika ada) dari testing set
print("\nNilai duplikat pada testing set:")
print(duplicate_rows_test)
```

## Melakukan Data cleaning pada data dengan nilai duplikat

```python
# Hapus nilai duplikat dari training set
train_set_cleaned = train_set.drop_duplicates()

# Hapus nilai duplikat dari testing set
test_set_cleaned = test_set.drop_duplicates()

# Periksa kembali jika ada nilai duplikat setelah pembersihan
duplicate_rows_train_cleaned = train_set_cleaned[train_set_cleaned.duplicated()]
duplicate_rows_test_cleaned = test_set_cleaned[test_set_cleaned.duplicated()]

# Tampilkan hasilnya
print("Jumlah nilai duplikat pada training set setelah data cleaning:")
print(len(duplicate_rows_train_cleaned))

print("\nJumlah nilai duplikat pada testing set setelah data cleaning:")
print(len(duplicate_rows_test_cleaned))
```

## Ganti tipe data attribute Weight jadi int

```python
# Mengubah tipe data kolom 'Weight' menjadi tipe data integer
train_set['Weight'] = train_set['Weight'].astype(int)

print(train_set['Weight'])
```

## Melakukan One Hot Encoding pada dataset

```python
# Lakukan one-hot encoding pada kolom kategorikal, misalnya 'Sport' dan 'Medal'
train_set_encoded = pd.get_dummies(train_set, columns=['Sport', 'Medal'])
test_set_encoded = pd.get_dummies(test_set, columns=['Sport', 'Medal'])

# Menampilkan beberapa baris pertama dari training set setelah one-hot encoding
print("Training Set setelah One-Hot Encoding:")
print(train_set_encoded.head())

# Menampilkan beberapa baris pertama dari testing set setelah one-hot encoding
print("\nTesting Set setelah One-Hot Encoding:")
print(test_set_encoded.head())
```
