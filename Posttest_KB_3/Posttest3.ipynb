## Mengimport Library dan Membaca Dataset

## Grafik Batang

## Berdasarkan Grafik Batang didapat data bahwa Atlet berusia 25-29 tahun lebih banyak dari Atlet yang berusia 20-24 pada Tahun 2000-2016

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataFrame = pd.read_csv('athlete_events.csv')

# Filter data tahun dan usia
filtered_data = dataFrame[(dataFrame['Year'] >= 2000) & (dataFrame['Year'] <= 2016) & (dataFrame['Age'] >= 20) & (dataFrame['Age'] <= 30)]

# Kelompokkan data berdasarkan tahun dan kelompok usia, lalu hitung jumlah atlet
grouped_data = filtered_data.groupby(['Year', pd.cut(filtered_data['Age'], bins=range(20, 31, 5))]).size().unstack(fill_value=0)

# Membuat grafik batang
grouped_data.plot(kind='bar', stacked=True, color=['#FF7F50', '#00CED1', '#228B22', '#8A2BE2'], figsize=(10,6))
plt.xlabel('Tahun')
plt.ylabel('Jumlah Atlet')
plt.title('Jumlah Atlet Berdasarkan Kelompok Usia dan Tahun')
plt.legend(title='Kelompok Usia', labels=['20-24', '25-29', '30-34'])
plt.show()
```

## Pie Chart

## Berdasarkan Pie Chart di dapat data persentase Atlet Pria 15.7% lebih banyak dari Atlet Wanita

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataFrame = pd.read_csv('athlete_events.csv')

# Filter data untuk tahun 2000-2016
filtered_data = dataFrame[(dataFrame['Year'] >= 2000) & (dataFrame['Year'] <= 2016)]

# Menghitung jumlah atlet pria dan wanita
gender_counts = filtered_data['Sex'].value_counts()

# Membuat pie chart
labels = ['Pria', 'Wanita']
colors = ['#ff9999','#66b3ff']
plt.pie(gender_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Atlet Pria dan Wanita (2000-2016)')
plt.axis('equal')  # Mengatur aspek lingkaran agar terlihat seperti lingkaran sempurna
plt.show()
```

## Heat Map

## Berdasarkan Heat Map dapat disimpulkan Semakin tua atlet, semakin besar kemungkinan tinngi dan berat badannya juga akan lebih besar

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
dataFrame = pd.read_csv('athlete_events.csv')

# Filter data untuk menghilangkan baris dengan nilai yang hilang
filtered_data = dataFrame[['Age', 'Height', 'Weight']].dropna()

# Menghitung matriks korelasi
correlation_matrix = filtered_data.corr()

# Membuat heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title('Heat Map Korelasi Antara Umur, Tinggi, dan Berat Atlet')
plt.show()
```

## Scatter Plot

## Berdasarkan Scatter Plot didapat data bahwa semakin tinggi seorang atlet, semakin besar kemungkinan berat badannya juga akan lebih besar

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataFrame = pd.read_csv('athlete_events.csv')

# Filter data untuk menghilangkan baris dengan nilai yang hilang
filtered_data = dataFrame[['Height', 'Weight']].dropna()

# Mengambil tinggi dan berat sebagai data x dan y
x = filtered_data['Height']
y = filtered_data['Weight']

# Membuat scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.5, color='blue')
plt.title('Scatter Plot Tinggi vs Berat')
plt.xlabel('Tinggi (cm)')
plt.ylabel('Berat (kg)')
plt.show()
```
